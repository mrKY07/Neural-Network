Question 1: What is Deep Learning? Briefly describe how it evolved and how it differs from traditional machine learning

Deep Learning is a branch of Artificial Intelligence that focuses on teaching machines to learn and make decisions using neural networks inspired by the human brain. Unlike traditional algorithms, deep learning systems can automatically discover patterns and features from raw data without much manual effort. It became popular due to the availability of large datasets, powerful GPUs, and better algorithms that made training deep neural networks practical.
The evolution of deep learning started with simple perceptrons in the 1950s, which could only solve basic problems. Later, advancements like backpropagation in the 1980s made multilayer networks possible, but progress was slow due to limited computing power. Around 2010, deep learning gained momentum with breakthroughs in computer vision and speech recognition, especially after ImageNet competitions showed its power.
Deep learning differs from traditional machine learning in how it handles features. Traditional ML requires manual feature engineering, meaning humans decide which attributes matter. In contrast, deep learning automatically extracts complex features layer by layer. For example, in image recognition, ML might use edges and shapes manually defined, while deep learning learns them directly from pixels.


Question 2: Explain the basic architecture and functioning of a Perceptron. What are its limitations?

A Perceptron is the simplest type of artificial neural network unit, introduced by Frank Rosenblatt in 1958. It mimics how a biological neuron works by taking multiple inputs, processing them, and producing a single output. The basic architecture consists of input nodes, weights, a summation function, an activation function, and an output.
Each input is assigned a weight, which indicates its importance. The perceptron multiplies each input by its weight and then adds them all together, often including a bias term to improve flexibility.
 Mathematically, this step can be written as:
 z = w1x1 + w2x2 + … + wnxn + b.
The summation result is then passed through an activation function (like step function, sigmoid, or ReLU) to decide the final output. In the original perceptron, a step function was used, producing either 0 or 1, making it suitable for binary classification tasks.
For example, if we train a perceptron to detect whether an email is spam, the inputs could be features like the presence of certain words, and the output would be 1 (spam) or 0 (not spam).
The perceptron learns by adjusting its weights during training using errors from wrong predictions, a process known as the perceptron learning rule. Though limited (it cannot solve non-linear problems like XOR), it is the foundation for modern deep learning architectures.
Limitations of Perceptron
Works well for simple cases but fails with non-linear tasks like the XOR problem.


Single-layer architecture ,it cannot capture complex patterns or higher-level abstractions needed in modern AI problems.


The step function prevents the use of gradient-based optimization methods like backpropagation.


Poor handling of noisy data – It struggles with overlapping classes or datasets that are not clearly separable.


Slow learning with large data – Weight adjustments are limited, and convergence can be difficult with big datasets.


Question 3: Describe the purpose of activation function in neural networks. Compare Sigmoid, ReLU, and Tanh functions.

The activation function is crucial in neural networks because it introduces non-linearity, allowing models to learn complex relationships. Among the most common functions are Sigmoid, Tanh, and ReLU, each with distinct properties and use cases.
The Sigmoid function maps inputs between 0 and 1, making it useful for probability-based tasks such as binary classification. However, it is not zero-centered, which can cause optimization difficulties. Its major drawback is the vanishing gradient problem, where very large or small inputs lead to gradients close to zero, slowing down learning in deep networks.
The Tanh function, on the other hand, maps inputs between -1 and 1, which makes it zero-centered. This allows both positive and negative values, helping balance the data during optimization. Compared to sigmoid, tanh often performs better in hidden layers. Still, it also suffers from the vanishing gradient problem when dealing with extreme input values.
The ReLU function is the most widely used today. It outputs the input directly if it is positive and returns zero if it is negative. ReLU is computationally efficient, avoids most vanishing gradient issues, and speeds up training in deep models. Its main limitation is the dying ReLU problem, where neurons that output only zero stop learning permanently.
In comparison, Sigmoid is suitable when probabilities are needed, Tanh is better for centered outputs in hidden layers, and ReLU dominates in modern deep learning because of its speed and effectiveness. Thus, while each function has strengths and weaknesses, ReLU remains the practical choice for most deep networks.


Question 4: What is the difference between Loss function and Cost function in neural networks? Provide examples.

In neural networks, the terms loss function and cost function are often used, and while they are related, there is a difference between them. Both deal with how far the model’s predictions are from the actual values, but they operate at different levels.
A loss function measures the error for a single training example. It tells us how much the model was wrong for one data point. For example, if the actual house price is ₹1,00,000 and the model predicts ₹90,000, the loss function calculates the difference between the two. Depending on the type of problem, different loss functions are used. For regression problems, Mean Squared Error (MSE) is commonly used, where the square of the difference between the actual and predicted value is taken. For classification problems, Cross-Entropy Loss is used to measure how close the predicted probabilities are to the actual class.
On the other hand, a cost function is the average of the loss values for the entire dataset. Instead of looking at just one example, it summarizes how well the model is performing across all data points. For example, if three samples have losses 1, 4, and 9, the cost function would be the average, i.e., (1+4+9)/3 = 4.67. The cost function is what the neural network tries to minimize during training using optimization algorithms like Gradient Descent.
The loss function tells us the error for one sample.


The cost function gives the overall error across all samples.


For instance, if we train a model to classify emails as spam or not spam, the loss function measures the error for one email, while the cost function gives the average error for the entire batch of emails.


Question 5: What is the role of optimizers in neural networks? Compare Gradient Descent, Adam, and RMSprop.

In neural networks, optimizers play the role of adjusting the weights and biases to reduce the overall error. They work together with the loss and cost functions. While the loss function tells us how wrong the model is, the optimizer decides how to update the weights so that the model improves in the next iteration. Without optimizers, training a neural network would be very slow or even impossible, especially with large datasets and deep architectures.
The most basic optimizer is Gradient Descent. It updates weights by moving in the direction of the negative gradient of the cost function. In simple terms, it takes small steps downhill until it reaches the lowest possible error. While effective, it can be slow for large datasets and may get stuck in local minima.

RMSprop (Root Mean Square Propagation) improves on this by adapting the learning rate for each parameter. It divides the learning rate by the average of recent squared gradients, which prevents large oscillations and makes training faster. RMSprop is especially useful for non-stationary problems like time-series data.

Adam (Adaptive Moment Estimation) combines the benefits of both Momentum and RMSprop. It keeps track of both the average of past gradients (like momentum) and the average of past squared gradients (like RMSprop). This makes Adam one of the most popular optimizers today because it is fast, efficient, and usually works well without much tuning.

Gradient Descent

This is the simplest optimizer that updates weights by moving in the opposite direction of the error slope.
It uses the same learning rate for all parameters, so choosing the right learning rate is very important.
Training can be slow on large datasets, and it can sometimes get stuck in wrong spots (local minima).
Good for simple tasks, but not suitable for very deep or complex networks.

RMSprop (Root Mean Square Propagation)

An improved version of Gradient Descent that gives each parameter its own adaptive learning rate.
It reduces big jumps in learning by keeping track of recent squared gradients, making the process smoother.
Works really well for problems like time-series or recurrent networks where data keeps changing.
Faster and more stable than plain Gradient Descent, but it does not use momentum, so it may not be the fastest overall.

Adam (Adaptive Moment Estimation)
Combines the best of both Momentum (smooth learning) and RMSprop (adaptive learning rate).
Tracks both the average of past gradients and squared gradients to make learning more accurate.
Does not require much fine-tuning of parameters and usually works well by default.
Very fast, efficient, and reliable, which is why it is the most widely used optimizer in modern deep learning.


```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Input data for AND gate
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])
y = np.array([[0],
              [0],
              [0],
              [1]])

# Building single-layer perceptron
model = Sequential()
model.add(Dense(1, input_dim=2, activation='sigmoid'))  # single neuron

# Compile model
model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'])

# Train model
history = model.fit(X, y, epochs=100, verbose=0)

# Predictions
predictions = model.predict(X)
print("Predictions:\n", np.round(predictions))

# Plot training loss
plt.plot(history.history['loss'])
plt.title('Training Loss over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

# Input values
x = np.linspace(-10, 10, 200)

# Compute outputs
y_sigmoid = sigmoid(x)
y_relu = relu(x)
y_tanh = tanh(x)

# Plot all activation functions
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.plot(x, y_sigmoid, 'b')
plt.title("Sigmoid Function")
plt.xlabel("x")
plt.ylabel("f(x)")

plt.subplot(1, 3, 2)
plt.plot(x, y_relu, 'r')
plt.title("ReLU Function")
plt.xlabel("x")
plt.ylabel("f(x)")

plt.subplot(1, 3, 3)
plt.plot(x, y_tanh, 'g')
plt.title("Tanh Function")
plt.xlabel("x")
plt.ylabel("f(x)")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical

# Load MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalize input
X_train = X_train / 255.0
X_test = X_test / 255.0

# One-hot encode labels
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Build multilayer neural network
model = Sequential([
    Flatten(input_shape=(28,28)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

# Compile model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train model
history = model.fit(X_train, y_train, epochs=5, batch_size=128, verbose=1)

# Print training accuracy
train_acc = history.history['accuracy'][-1]
print(f"Training Accuracy: {train_acc*100:.2f}%")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.utils import to_categorical

# Load Fashion MNIST dataset
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

# Normalize input
X_train = X_train / 255.0
X_test = X_test / 255.0

# One-hot encode labels
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Build simple neural network
model = Sequential([
    Flatten(input_shape=(28,28)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

# Compile model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train model
history = model.fit(X_train, y_train, epochs=10, batch_size=128, validation_data=(X_test, y_test), verbose=1)

# Plot loss curves
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss Curves')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

# Plot accuracy curves
plt.subplot(1,2,2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy Curves')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.metrics import classification_report, roc_auc_score
from imblearn.over_sampling import SMOTE
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt

# Example: Generating synthetic dataset
np.random.seed(0)
n_samples = 10000
# Features: transaction_amount, merchant_id, customer_location
X_numeric = np.random.rand(n_samples, 1) * 1000  # transaction amount
X_categorical = np.random.randint(0, 10, size=(n_samples, 2))  # merchant_id, location
y = np.random.choice([0,1], size=(n_samples,), p=[0.95,0.05])  # imbalanced target

# Preprocessing
X = np.hstack([X_numeric, X_categorical])

# Scale numeric features
scaler = StandardScaler()
X[:,0] = scaler.fit_transform(X[:,0].reshape(-1,1)).flatten()

# Encode categorical features (updated for latest scikit-learn)
encoder = OneHotEncoder(sparse_output=False)
X_cat_encoded = encoder.fit_transform(X[:,1:])
X_processed = np.hstack([X[:,0:1], X_cat_encoded])

# Handle class imbalance using SMOTE
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X_processed, y)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42)

# Build multilayer neural network
model = Sequential([
    Dense(128, activation='relu', input_shape=(X_train.shape[1],)),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.2),
    Dense(1, activation='sigmoid')  # binary output
])

# Compile model
model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])

# Train model
history = model.fit(X_train, y_train, epochs=20, batch_size=64, validation_split=0.2, verbose=1)

# Evaluate model
y_pred_prob = model.predict(X_test).flatten()
y_pred = (y_pred_prob > 0.5).astype(int)

print("Classification Report:\n", classification_report(y_test, y_pred))
print("ROC-AUC Score:", roc_auc_score(y_test, y_pred_prob))

# Plot loss and accuracy curves
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss Curve')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy Curve')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.show()
```
